```python
#IMPORTING THE DATASET
df = pd.read_csv("/content/drive/MyDrive/ML assignment/iris.csv",header=None,
                 names=['sepal length','sepal width', 'petal length','petal width','target'])
```

```python
## Taking all the numeric values into the consideration
X = df.iloc[:,:-1]
X.dtypes
# As here the numerical columns are converted into object types so converting it into numeric
```

```python
X = X.astype(float)
X.dtypes
```

## Let's perform PCA on the dataset

```python
## importing the necessary libraries
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# performing StandardScaler on numeric values
scaler = StandardScaler()
x = scaler.fit(X)
```

```python
# Transforming the values
x = scaler.transform(X)
x
```

```python
import seaborn as sns

covariance_matrix = np.cov(x, rowvar=False)


plt.figure(figsize=(4, 4))
sns.heatmap(covariance_matrix, annot=True, cmap='viridis', fmt=".2f",
            xticklabels=x.columns, yticklabels=x.columns)
plt.title('Covariance Matrix')
plt.show()
```

```python
features = ['sepal length','sepal width','petal length','petal width']
x = pd.DataFrame(data = x, columns = features)
x.head(10)
```

```python
y = df.loc[:,['target']].values
y
```

```python
## PCA projection to 2D
pca = PCA(n_components=2)
pc = pca.fit_transform(x)
pca_df = pd.DataFrame(data = pc
              , columns = ['pca1','pca2'])
pca_df.head(10)
```

```python
df[['target']].head()
```

```python
pca_df_final = pd.concat([pca_df,df[['target']]],axis=1)
pca_df_final.head(10)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('pca1', fontsize = 15)
ax.set_ylabel('pca2', fontsize = 15)
ax.set_title('2D PCA', fontsize = 20)


targets = ['Setosa', 'Versicolor', 'Virginica']
colors = ['r', 'g', 'b']
for species, color in zip(targets,colors):
    indicesToKeep = pca_df_final['target'] == species
    ax.scatter(pca_df_final.loc[indicesToKeep, 'pca1']
               , pca_df_final.loc[indicesToKeep, 'pca2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
